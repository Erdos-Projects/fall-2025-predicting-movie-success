```python
import csv
import pandas as pd
import numpy as np
```

```python
non_awarded_results = 'GitHub/fall-2025-predicting-movie-success/source_data/nonwinner_reviews_all_times.csv'
non_winners_df = pd.read_csv(non_awarded_results)

non_winners_df = non_winners_df.drop_duplicates()

non_winners_df.loc[-1] = non_winners_df.columns
non_winners_df.index = non_winners_df.index + 1  # shifting index
non_winners_df.sort_index(inplace=True) 

non_winners_df.columns = ['Movie', '0.5', '1', '1.5', '2', '2.5', '3', '3.5', '4', '4.5', '5']

non_winners_df.set_index('Movie', inplace=True)


```

```python
cols = ['0.5', '1', '1.5', '2', '2.5', '3', '3.5', '4', '4.5', '5']

for col in cols:
    non_winners_df[col] = non_winners_df[col].astype(int)
```

```python
non_winners_df['total'] = (non_winners_df['0.5'] + non_winners_df['1'] + non_winners_df['1.5'] + non_winners_df['2'] + non_winners_df['2.5'] + non_winners_df['3'] + non_winners_df['3.5'] + non_winners_df['4'] + non_winners_df['4.5'] + non_winners_df['5'])
non_winners_df['mean'] = (non_winners_df['0.5']*0.5 + non_winners_df['1']*1.0 + non_winners_df['1.5']*1.5 + non_winners_df['2']*2.0 + non_winners_df['2.5']*2.5 + non_winners_df['3']*3.0 + non_winners_df['3.5']*3.5 + non_winners_df['4']*4.0 + non_winners_df['4.5']*4.5 + non_winners_df['5']*5.0)/non_winners_df['total']
non_winners_df['mean'] = non_winners_df['mean'].round(2)
```

```python
movies_2025 = ['dune-part-two', 'a-complete-unknown', 'the-substance',
       'furiosa-a-mad-max-saga', 'a-real-pain']
movies_2024 = ['the-color-purple', 'maestro-2023', 'past-lives', 'the-zone-of-interest']
movies_2023 = ['tar-2022', 'the-menu-2022', 'bardo-false-chronicle-of-a-handful-of-truths',  'babylon-2022', 'rrr']
movies_2022 = ['nightmare-alley-2021', 'belfast', 'being-the-ricardos','house-of-gucci', 'no-time-to-die-2021',
       'the-french-dispatch']
movies_2021 = [ 'the-father-2020', 'the-mauritanian', 'minari',
       'judas-and-the-black-messiah', 'one-night-in-miami', 'sound-of-metal']
movies_2020 = ['ford-v-ferrari', 'the-two-popes', 'knives-out-2019', 'joker-2019']
movies_2019 = [ 'vice-2018', 'a-star-is-born-2018', 'crazy-rich-asians',
       'if-beale-street-could-talk', 'blackkklansman', 'deadpool-2',
       'mary-poppins-returns', 'a-quiet-place-2018']
movies_2018 = ['darkest-hour',
       'phantom-thread', 'the-post-2017', 'the-big-sick',
       'the-florida-project', 'the-disaster-artist', 'logan-2017']
movies_2017 = ['hacksaw-ridge', 'manchester-by-the-sea', 'sing-street', 'sully',
       'hell-or-high-water', 'nocturnal-animals', 'the-lobster', 'deadpool',
       'the-jungle-book-2016', 'captain-fantastic']
movies_2016 = ['ant-man', 'carol-2015',
       'steve-jobs', 'star-wars-the-force-awakens']

movies = [movies_2025, movies_2024, movies_2023, movies_2022, movies_2021, movies_2020, movies_2019, movies_2018, movies_2017, movies_2016] 

oscar_winners = ["anora", "oppenheimer-2023", "everything-everywhere-all-at-once", 
                 "coda-2021", "nomadland", "parasite-2019", "green-book", "the-shape-of-water",
                "moonlight-2016", "spotlight" ]
```

```python
data = []
columns = ['Winner Average', 'Nominated Non-Winner Averages']
index = [str(num) for num in range(2025, 2015,-1)] 
df = pd.DataFrame(data, columns=columns, index=index)
```

```python
for i in range(2025, 2015,-1):
    tot = 0
    for movie in movies[2025-i]:
        tot += non_winners_df.loc[movie, 'mean']
    avg = tot/len(movies[2025-i])
    df.loc[str(i), 'Nominated Non-Winner Averages'] = avg.round(2)
```

```python
file_path = 'GitHub/fall-2025-predicting-movie-success/modified_data/connor_modified_data/movie_reviews_all.csv'
winners_df = pd.read_csv(file_path)

winners_df = winners_df.drop_duplicates()
winners_df.set_index('moviename', inplace=True)

for index, value in enumerate(oscar_winners):
    df.loc[str(2025-index), 'Winner Average'] = winners_df.loc[value, 'mean']

df['Difference'] = df['Winner Average'] - df['Nominated Non-Winner Averages']
```
