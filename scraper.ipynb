```python
import requests
from bs4 import BeautifulSoup
import time
import random
import csv


user = 'jay'
diary_page = 1
records = []
while True:
    print(diary_page)
    url = f'https://letterboxd.com/{user}/films/diary/page/{diary_page}' 
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'lxml')


    rows = soup.select('tr.diary-entry-row.viewing-poster-container[data-object-name="review"]')
    if not rows:
        break
    for r in rows:
        film = r.select_one('[data-item-link]')['data-item-link'][6:-1] #name of the film

        date_href = r.select_one('.col-daydate a.daydate').get('href', '')
        date = date_href[date_href.find('for/')+4:] #date of review

        rating_raw = r.select_one('.col-rating .rating').get_text()
        rating = rating_raw.count('★') + rating_raw.count('½')/2 #review
        records.append({
            "username": user,
            "film": film,
            "date": date,
            "rating": rating,
        })
    diary_page += 1
    time.sleep(45)
with open("letterboxd_diary.csv", "a", newline="", encoding="utf-8") as f:
            fieldnames = ["username", "film", "date", "rating"]
            w = csv.DictWriter(f, fieldnames=fieldnames)
            w.writerows(records)




```

```python
import csv
from itertools import islice
import requests
from bs4 import BeautifulSoup
import time
import random
import csv

start=3064
end=3269
with open("popular_users_ratings.csv", "a", encoding="utf-8") as popular_users_ratings,\
    open("popular_users.csv", "r", encoding="utf-8") as popular_users:
    
    reader = csv.reader(popular_users)
    writer = csv.writer(popular_users_ratings)
    for row in islice(reader, start, end):
        try:
            user = row[0]
            #ratings = [username, #0.5 star, #1 star, ..., #5 star, total]
            ratings = [user]
            url = f'https://letterboxd.com/{user}'
            page = requests.get(url)
            soup = BeautifulSoup(page.text, 'lxml')

            # total count might not be integer due to formatting, eg. 10.5k, for this reason i will
            # just compute the sum
            # total_rating = int(soup.select('.section.ratings-histogram-chart>.all-link')[0].get_text().replace(',',''))

            rating_bars = soup.select('div.rating-histogram ul > li.rating-histogram-bar')
            compact = (len(rating_bars) == 5)

            for rating_bar in rating_bars:
                if compact:
                    try:
                        ratings.extend([0, int(rating_bar.get_text().split()[0].replace(',', ''))])
                    except IndexError:
                        ratings.extend([0, 0])
                else:
                    try:
                        ratings.append(int(rating_bar.get_text().split()[0].replace(',', '')))
                    except IndexError:
                        ratings.append(0)

            total_rating = sum(ratings[1:])
            ratings.append(total_rating)
            print(f'{ratings}')
            writer.writerow(ratings)
            popular_users_ratings.flush()
            time.sleep(31)
        except Exception as e:
            #extra nap time idk
            print(f'edge case {user}')
            time.sleep(45)
```
